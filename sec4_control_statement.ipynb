# **Python Flow Control Statements**

Flow control statements allow a program to decide the sequence in which code executes. Python mainly has conditional statements and loop statements.

#### **1. Conditional Statements**

Conditional statements allow you to execute code only if certain conditions are true.

a) if Statement

```python
x = 10
if x > 0:
    print("x is positive")
```

b) if-else Statement

```python
x = -5
if x > 0:
    print("x is positive")
else:
    print("x is not positive")
```

c) if-elif-else Statement

```python
x = 0
if x > 0:
    print("x is positive")
elif x == 0:
    print("x is zero")
else:
    print("x is negative")
```

```python
# check a number is odd or even?

num = int(input("Enter a number: "))
if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

2. Nested if Statement

```python
#positive and even?
num = int(input("Enter a number: "))

if num > 0 and num % 2 == 0:
    print("The number is positive and even")
else:
    print("The number is NOT positive and even")
```

find the sum of numbers from 1 to 100

```python
# Initialize sum
sum = 0

# Loop from 1 to 100
for i in range(1, 101):
   sum=sum+i # Add each number to total

# Print the result
print("The sum of natural  numbers from 1 to 100 is:", sum)
```

find the sum of even numbers from 1 to 50

```python
# Initialize sum
total = 0

# Loop through numbers from 1 to 50
for i in range(1, 51):
    if i % 2 == 0:   # Check if the number is even
        total += i   # Add it to the total

# Print the result
print("The sum of even numbers from 1 to 50 is:", total)
```
