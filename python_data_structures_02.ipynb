# **Python Data Types and Data Structures**

## 1. Data Types
Data types represent the **type of a single value**.

| Type   | Example          | Description               |
|--------|-----------------|---------------------------|
| int    | `x = 5`         | Integer (whole number)    |
| float  | `y = 5.6`       | Decimal number            |
| str    | `name = "Swathi"`| String (text)            |
| bool   | `a = True`      | Boolean (True/False)      |
| None   | `x = None`      | No value                  |

# **List Index in Python**

A list stores multiple values. Each value has a **position** called an **index**.

## Key Points
- **Index starts from 0** (first element → index 0)  
- **Negative index** counts from the end (`-1` is last element)


````python
fruits = ["apple", "banana", "mango", "orange"]
fruits[-4]
````

````python
fruits[2]
````

````python
fruits[-1]
````

````python
words = ["hi", "hello", "happy", "dude","happy", "python", "code"]
words.index("happy") #position
````

# List Indexing with Step in Python

Python slicing syntax:

# **list[start : stop : step]**

* start → index to start from (inclusive)

* stop → index to go up to (exclusive)

* step → how many indices to jump


````python

words = ["hi", "hello", "happy", "dude", "python", "code"]

words[1:]
````

````python
words[1::2]
````
