```python
# Import pandas
import pandas as pd

# Create a dictionary of students
students = {
    "Name": ["Alex", "Brian", "Chloe"],
    "Age": [16, 17, 16],
    "Grade": ["A", "B", "A"],
}

# Create DataFrame
df = pd.DataFrame(students)

# Display the DataFrame
df
```

```python
# 1. keys() - get all keys
print("\nKeys:", students.keys())
```

```python
# 2. values() - get all values
print("\nValues:", students.values())
```

```python
# 3. items() - get key-value pairs
print("\nItems:", students.items())
```

```python
# 4. get() - access value of a key
print("\nGet 'Name':", students.get("Name"))
print("Get 'City' (not present):", students.get("City"))
```

```python
# 5. update() - add or update key-value pair
students.update({"Age": [16, 17, 18]})      # update Age
students.update({"City": ["NY", "LA", "SF"]})  # add new key
print("\nUpdated Dictionary:", students)
```

```python
# 6. pop() - remove a key
popped_value = students.pop("Grade")
print("\nPopped 'Grade':", popped_value)
print("Dictionary after pop:", students)
```

```python
# Original dictionary
students = {
    "Name": ["Alex", "Brian", "Chloe"],
    "Age": [16, 17, 16],
    "Grade": ["A", "B", "A"],
}

# Find the index of Brian
index = students["Name"].index("Brian")

# Update Brian's age
students["Age"][index] = 20

print(students)
```

```python
#remove Chloeâ€™s data- Find the index of Chloe
index = students["Name"].index("Chloe")

# Remove Chloe's data from all lists
for key in students:
    students[key].pop(index)

print(students)
```
