# 1. Python IDEs (Integrated Development Environments)

Python code can be written and executed in different environments:

## a) IDLE (Python’s default IDE)
- Comes installed with Python.
- Simple interface: `File → New File → Write code → Run`.
- Run a program: Press **F5** or `Run → Run Module`.
- Good for beginners.

## b) Anaconda Environment (Jupyter Notebook)

https://repo.anaconda.com/archive/Anaconda3-2024.10-1-Windows-x86_64.exe

- Popular for data science and AI.
- Allows interactive coding (code + output in the same notebook).
- Supports **Markdown** for explanations.
- Run a cell: **Shift + Enter**.

## c) Online IDEs
- Examples: Google Colab, Replit, Kaggle Notebooks.
- No installation required; run code in your browser.
- Useful for sharing projects and collaborating.


# 2. Variables in Python

Variables are used to store values. You don’t need to declare a type explicitly; Python infers it automatically.




```python
## Assigning values to variables

x = 10           # integer
type(x)
```

```python
name = "Swathi"  # string
type(name)
```

```python
price = 99.99    # float
type(price)
```

```python
Z= True # boolean
type(Z)
```


### Rules for variable names:

* Must start with a letter or underscore _.

* Cannot start with a number.

* No spaces or special characters (except _).

* Case-sensitive (Age ≠ age).

# 4. Operators in Python

Operators are used to perform operations on variables and values.

## a) Arithmetic Operators

| Operator | Name         | Example  | Output |
|----------|-------------|---------|--------|
| `+`      | Addition     | 5 + 3   | 8      |
| `-`      | Subtraction  | 5 - 3   | 2      |
| `*`      | Multiplication | 5 * 3 | 15     |
| `/`      | Division     | 5 / 2   | 2.5    |
| `//`     | Floor Division | 5 // 2 | 2      |
| `%`      | Modulus      | 5 % 2   | 1      |
| `**`     | Exponent     | 5 ** 2  | 25     |

## b) Comparison Operators

| Operator | Name             | Example  | Output |
|----------|-----------------|---------|--------|
| `==`     | Equal to        | 5 == 5  | True   |
| `!=`     | Not equal to    | 5 != 3  | True   |
| `>`      | Greater than    | 5 > 3   | True   |
| `<`      | Less than       | 5 < 3   | False  |
| `>=`     | Greater or equal | 5 >= 5 | True   |
| `<=`     | Less or equal   | 5 <= 4  | False  |



```python
# Comparison Operators Examples

x = 7
y = 9

print(x == y)

```

```python
print(x != y)
```

```python
print(x > y)
```

```python
print(x < y)
```

```python
print(x >= 7)
```

```python
print(y <= 10)
```

## c) Logical Operators

| Operator | Name    | Example         | Output |
|----------|--------|----------------|--------|
| `and`    | AND     | True and False  | False  |
| `or`     | OR      | True or False   | True   |
| `not`    | NOT     | not True        | False  |



```python
# Logical Operators Example with numbers

x = 7
y = 9

# AND operator
print(x > 5 and y < 10)  # True, both conditions are True


```

```python
# OR operator
print(x > 5 or y > 10)   # True, at least one condition is True

```

```python
# NOT operator
print(not(x > 5))        # False, reverses the condition
```

## d) Assignment Operators

| Operator | Name            | Example   | Output       |
|----------|----------------|----------|--------------|
| `=`      | Assign          | x = 5    | 5            |
| `+=`     | Add and assign  | x += 3   | x = x + 3    |
| `-=`     | Subtract and assign | x -= 2 | x = x - 2  |
| `*=`     | Multiply and assign | x *= 4 | x = x * 4  |
| `/=`     | Divide and assign | x /= 2 | x = x / 2   |


```python
# Assignment Operators Examples

x = 5
print(x)


```

```python
x += 3
print(x)

```

```python
x -= 2
print(x)

```

```python
x *= 4
print(x)

```

```python
x /= 2
print(x)
```

# Key Features of Python

1. **Easy to Learn and Use**
   - Simple, readable syntax similar to English.
   - Great for beginners and professionals.

2. **Interpreted Language**
   - Code is executed line by line.
   - Easier debugging and faster development.

3. **Dynamically Typed**
   - No need to declare data types (Python decides automatically).

4. **High-Level Language**
   - You don’t need to manage memory or hardware-level details.

5. **Extensive Libraries**
   - Rich built-in modules + external libraries like NumPy, Pandas, TensorFlow, etc.

6. **Object-Oriented**
   - Supports classes, objects, inheritance, etc.

7. **Open Source**
   - Free to use, modify, and distribute.

8. **Supports Multiple Programming Paradigms**
    - Procedural
    - Object-oriented
    - Functional programming

9. **Great for AI, Machine Learning, and Data Science**
    - Most popular language in AI/ML because of libraries and simplicity.
